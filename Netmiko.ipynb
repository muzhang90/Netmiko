From the source code of send_command() in BaseConnection:

if use_textfsm:
            structured_output = get_structured_data(
                output,
                platform=self.device_type,
                command=command_string.strip(),
                template=textfsm_template,
            )
            # If we have structured data; return it.
            if not isinstance(structured_output, str):
                return structured_output


def get_structured_data(raw_output, platform=None, command=None, template=None):
    """
    Convert raw CLI output to structured data using TextFSM template.
    You can use a straight TextFSM file i.e. specify "template". If no template is specified,
    then you must use an CliTable index file.
    """
    if platform is None or command is None:
        attrs = {}
    else:
        attrs = {"Command": command, "Platform": platform}

    if template is None:
        if attrs == {}:
            raise ValueError(
                "Either 'platform/command' or 'template' must be specified."
            )
        template_dir = get_template_dir()
        index_file = os.path.join(template_dir, "index")
        textfsm_obj = clitable.CliTable(index_file, template_dir)
        return _textfsm_parse(textfsm_obj, raw_output, attrs)


It is done via Textfsm class CliTable:

Mapping the vendor name, command, hostname, and template together.

# Create your own Template,Vendor,command and hostname mapping

```python
from textfsm import clitable
from netmiko import ConnectHandler
from pprint import pprint



juniper_qfx10k2={
    'device_type': 'juniper_junos',
    'host':   '10.85.168.43',
    'username': 'labroot',
    'password': 'lab123',

}

qfx_qfx10k2_connection=netmiko.ConnectHandler(**juniper_qfx10k2)
qfx_show_version=qfx_qfx10k2_connection.send_command('show version')

print(qfx_show_version)

cli_table=clitable.CliTable('index','.')
attr={'Hostname': 'qfx10k2', 'Command': 'show version' , 'Vendor': 'Juniper'}
cli_table.ParseCmd(qfx_show_version,attr)
print(cli_table)
```

# Use Netmiko to do file transfer

```python
from netmiko import ConnectHandler,file_transfer

juniper_qfx10k2={
    'device_type': 'juniper_junos',
    'host':   '10.85.168.43',
    'username': 'labroot',
    'password': 'lab123',

}

qfx_qfx10k2_connection=netmiko.ConnectHandler(**juniper_qfx10k2)

transfer_result=file_transfer(qfx_qfx10k2_connection,source_file='output.txt',dest_file='output.txt',file_system='/var/tmp',direction='put',overwrite_file=True)
print(transfer_result)
```
